```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.offline as pyo

csv_path = "C:/Users/djgoo/DataAnalysis/Project-3/EV Population Data/Electric_Vehicle_Population_Data.csv"
```

```python
EV_df = pd.read_csv(csv_path)
EV_df
```

```python
columns = ["VIN (1-10)", "Make", "Model", "Model Year", "Electric Range", "Electric Utility"]
EV_df_Range = EV_df[columns]
EV_df_Range
```

```python
EV_df_Range = EV_df_Range[EV_df_Range["Electric Range"] !=0]
EV_df_Range
```

```python
plt.scatter(EV_df_Range['Model Year'], EV_df_Range['Electric Range'])
plt.xlabel('Year')
plt.ylabel('Electric Range')
plt.title('Electric Vehicle Range Over Years')
plt.show()
```

```python
plt.bar(EV_df_Range['Model'], EV_df_Range['Electric Range'])
plt.xlabel('Year')
plt.xticks(rotation = 45, ha='right')
plt.ylabel('Electric Range')
plt.title('Electric Vehicle Range By Model')
plt.show()
```

```python
EV_range_mean = EV_df_Range.groupby(['Make','Model', 'Model Year'])['Electric Range'].mean().reset_index()
EV_range_mean
```

```python
unique_makes = EV_range_mean['Make'].unique()

# Create a separate line graph for each Make
for make in unique_makes:
    make_data = EV_range_mean[EV_range_mean['Make'] == make]
    
    # Plot a line for each corresponding model
    for model in make_data['Model'].unique():
        model_data = make_data[make_data['Model'] == model]
        plt.plot(model_data['Model Year'], model_data['Electric Range'], label=f"{make} - {model}")

# Customize the plot
plt.xlabel('Model Year')
plt.ylabel('Electric Range')
plt.title('Electric Vehicle Range by Model Through The Years')
plt.legend(title='Make - Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
Linedf = EV_range_mean.pivot(index='Model Year', columns='Model', values='Electric Range')
Linedf.plot(kind='line', stacked=False, figsize=(12, 6))
plt.xlabel('Year')
plt.ylabel('Electric Range')
plt.title('Washington State Electric Vehicle Range by Model Through The Years')
plt.legend(title='Model', bbox_to_anchor=(1.02, 1.02), loc='upper left')
plt.show() 
```

```python
fig = px.line(EV_range_mean, x='Model Year', y='Electric Range', color='Model',
              labels={'Electric Range': 'Electric Range', 'Model Year': 'Year'}, height=600,
              line_group='Model', hover_name='Model')

fig.update_layout(title='Electric Vehicle Range by Model Through The Years',
                  legend_title='Model',
                  xaxis_title='Year',
                  yaxis_title='Electric Range')

fig.show()
```

```python
Linechart = px.line(EV_range_mean, x='Model Year', y='Electric Range', color='Model',
              labels={'Electric Range': 'Electric Range', 'Model Year': 'Model Year'}, height=600,
              line_group ='Make', line_dash='Make', hover_name='Model')

fig.update_layout(title='Washington State Electric Vehicle Range by Model Through The Years',
                  legend_title='Model, Make',
                  xaxis_title='Year',
                  yaxis_title='Electric Range')

fig.show()
```

```python
filepathlinechart = "Linechart.html"
pyo.plot(Linechart, filename=filepathlinechart, auto_open=False)
```
