```python
import matplotlib.pyplot as plt

# List of makes and their corresponding counts
makes = [
    'TESLA', 'CHRYSLER', 'KIA', 'NISSAN', 'MINI', 'TOYOTA', 'FORD', 'VOLVO', 'CHEVROLET',
    'FIAT', 'AUDI', 'BMW', 'JEEP', 'POLESTAR', 'HONDA', 'VOLKSWAGEN', 'MERCEDES-BENZ',
    'RIVIAN', 'PORSCHE', 'HYUNDAI', 'SUBARU', 'SMART', 'JAGUAR', 'MITSUBISHI', 'CADILLAC',
    'LEXUS', 'ALFA ROMEO', 'MAZDA', 'GENESIS', 'LUCID', 'LAND ROVER', 'DODGE', 'TH!NK',
    'WHEEGO ELECTRIC CARS', 'FISKER', 'BENTLEY', 'AZURE DYNAMICS'
]

counts = [
    tesla_count, chrysler_count, kia_count, nissan_count, mini_count, toyota_count, ford_count,
    volvo_count, chevy_count, fiat_count, audi_count, bmw_count, jeep_count, pole_count, honda_count,
    volks_count, benz_count, rivian_count, porsche_count, hyundai_count, subaru_count, smart_count,
    jaguar_count, mitsubishi_count, cadillac_count, lexus_count, alfa_count, mazda_count, genesis_count,
    lucid_count, rover_count, dodge_count, think_count, wheego_count, fisker_count, bentley_count,
    azure_count
]

# Create a bar chart
plt.figure(figsize=(14, 8))
plt.bar(makes, counts, color='skyblue')
plt.title('Electric Vehicle Counts by Make as of 2023')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Filter the DataFrame for the year 2019 and Tesla make
tesla_2019 = df[(df['Model Year'] == 2019) & (df['Make'] == 'TESLA')]

# Count the occurrences of Tesla in 2019
tesla_2019_count = len(tesla_2019)

# Create a bar chart for Tesla in 2019
plt.figure(figsize=(6, 4))
plt.bar(['TESLA'], [tesla_2019_count], color='blue')
plt.title('Tesla Vehicles in 2019')
plt.xlabel('Make')
plt.ylabel('Count')
plt.show()
```

```python

# Specify the makes of interest
makes_of_interest = ['TESLA', 'NISSAN', 'CHEVROLET', 'FORD']

# Filter the DataFrame for the year 2019 and the specified makes
filtered_data = df[(df['Model Year'] == 2019) & (df['Make'].isin(makes_of_interest))]

# Count the occurrences of each make in 2019
make_counts = filtered_data['Make'].value_counts()

# Create a bar chart for the selected makes in 2019
plt.figure(figsize=(10, 6))
plt.bar(make_counts.index, make_counts.values, color=['blue', 'green', 'red', 'orange'])  # Adjust colors as needed
plt.title('Electric Vehicles by Make in 2019')
plt.xlabel('Make')
plt.ylabel('Count')
plt.show()
```
